```python
def split_text_with_overlap(text, chunk_size, overlap_words):
    words = text.split()
    chunks = []
    
    i = 0
    while i < len(words):
        start_idx = i
        end_idx = i + chunk_size
        overlap_idx = end_idx - overlap_words
        
        if overlap_idx < i:
            overlap_idx = i

        txt = ' '.join(words[start_idx:end_idx])
        chunks.append( txt )
        print (len(txt), start_idx, end_idx )
        i = overlap_idx

    return chunks

df_chunks_ramayan = pd.DataFrame(columns=['Section_Id','Chunk_Id','Chunk'])

path = r"D:\github-blog\dasarpai\_corpus\Ramayana-ManmathNath"
files= os.listdir(path)

# loop for all Kandas 
for bookno in range(1,8):
    filename = path + "\\" +files[bookno-1] 

    print(bookno, filename)
    # f"D:\github-blog\dasarpai\_corpus\Ramayana-ManmathNath\Ramayana-{bookno}-Balakanda.md"

    #open kanda file
    with open(filename, "r") as f:
        all_lines = f.readlines(0)
        f.close()
    
    #get all the sections of that kanda
    section_text =  {}
    text = []
    section_no = 0
    for i, line in enumerate(all_lines): 
        # print(line)
        if  line.lstrip()[0:10]=="## SECTION":
            section_text[section_no] = " ".join(text)
            section_no += 1
            
            text = []
        text.append(line)

    # remove first section because it is md file header.
    section_text.pop(0)

    #create chunk for section and add in df_chunk 
    for section_id,text in section_text.items():
        # chunks = text_splitter.split_text(text[3:])
        chunks = split_text_with_overlap(text[3:],1000,100)

        for chunk_id, chunk in enumerate(chunks): 
            section_id1 =  "Book"+str(bookno)+"_"+str(section_id)+"_"+str(chunk_id+1)
            new_row = section_id1, chunk_id +1, chunk 
            # print(chunk)
            df_chunks_ramayan.loc[section_id1] = new_row
        # if section_id == 4:
        # 	break 
```
