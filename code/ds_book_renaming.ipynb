```python
df = pd.read_csv('ds-books.csv')
```

```python
df['Type'] = df['Book'].str.split('-').str[-1].str[:-4].str.upper()
```

```python
df['X'] = df['Book'].str.split('-').str[0:-1]
```

```python
# Function to concatenate values
def concatenate_values(row):
    return '-'.join(map(str, row['X'])) + '-' +row['Type'] + '.pdf'

# Apply the function to create the new column Z
df['New_Filename'] = df.apply(lambda row: concatenate_values(row), axis=1)
```
