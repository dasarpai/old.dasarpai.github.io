```python
import pdfplumber

def extract_text_from_pdf(input_pdf, output_text):
    # Open the PDF file
    with pdfplumber.open(input_pdf) as pdf:
        # Initialize an empty string to store extracted text
        text_content = ""

        # Iterate through each page in the PDF
        for page_number in range(len(pdf.pages)):
            # Get the text of the current page
            page = pdf.pages[page_number]
            text_content += page.extract_text()

        # Write the extracted text to a text file with explicit UTF-8 encoding
        with open(output_text, "w", encoding="utf-8") as text_file:
            text_file.write(text_content)

# # Example usage
# input_pdf = "your_input.pdf"
# output_text = "output.txt"
# extract_text_from_pdf(input_pdf, output_text)
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(input_pdf, output_text):
    # Open the PDF file
    pdf_document = fitz.open(input_pdf)

    # Initialize an empty string to store extracted text
    text_content = ""

    # Iterate through each page in the PDF
    for page_number in range(pdf_document.page_count):
        # Get the text of the current page
        page = pdf_document.load_page(page_number)
        text_content += page.get_text()

    # Close the PDF document
    pdf_document.close()

    # Write the extracted text to a text file with explicit UTF-8 encoding
    with open(output_text, "w", encoding="utf-8") as text_file:
        text_file.write(text_content)

# # Example usage
# input_pdf = "your_input.pdf"
# output_text = "output.txt"
# extract_text_from_pdf(input_pdf, output_text)
```

```python
from PIL import Image
import pytesseract
from pdf2image import convert_from_path

def extract_text_from_pdf(input_pdf, output_text):
    # Convert each PDF page to an image
    images = convert_from_path(input_pdf, 500)

    # Initialize an empty string to store extracted text
    text_content = ""

    # Iterate through each image and perform OCR
    for i, image in enumerate(images):
        # Save the image to a temporary file
        image_path = f"page_{i + 1}.png"
        image.save(image_path, "PNG")

        # Perform OCR on the image
        page_text = pytesseract.image_to_string(Image.open(image_path), lang="eng+deva")

        # Append the extracted text to the result
        text_content += page_text

    # Write the extracted text to a text file with explicit UTF-8 encoding
    with open(output_text, "w", encoding="utf-8") as text_file:
        text_file.write(text_content)

# # Example usage
# input_pdf = "your_input.pdf"
# output_text = "output.txt"
# extract_text_from_pdf(input_pdf, output_text)
```
