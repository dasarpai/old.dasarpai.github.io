```python
import os

def replace_layout_in_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    with open(file_path, 'w', encoding='utf-8') as file:
        for line in lines:
            if line.strip().startswith('layout:'):
                file.write('layout: wiapost-layout\n')
            else:
                file.write(line)

def process_files_in_directory(directory_path):
    for root, dirs, files in os.walk(directory_path):
        for file_name in files:
            if file_name.endswith(('.md', '.html', '.yml')):  # Adjust file extensions if needed
                file_path = os.path.join(root, file_name)
                replace_layout_in_file(file_path)
                print(f'Processed file: {file_path}')

if __name__ == "__main__":
    directory_path = 'D:\github-blog\mm.github.io\_wiapost'  # Replace with your folder path
    process_files_in_directory(directory_path)
```
