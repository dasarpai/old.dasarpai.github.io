```python
import requests

# Replace with your GitHub username
username = "dasarpai"

# GitHub API URL for listing starred repositories (initial page)
url = f"https://api.github.com/users/{username}/starred"

# Initialize variables for pagination
all_repos = []
page = 1

while True:
    # Send GET request to the GitHub API for the current page
    response = requests.get(url, params={'page': page, 'per_page': 100})

    # Check if the request was successful
    if response.status_code == 200:
        repos = response.json()

        # Break if no more repositories are returned
        if not repos:
            break

        # Add the current page of repos to the full list
        all_repos.extend(repos)

        # Increment page number for the next request
        page += 1
    else:
        print(f"Failed to fetch starred repositories. Status Code: {response.status_code}")
        break

# Print table headers
print(f"Sno. | Repo Name | Url | Repo Description | Language | Starred | Fork")
print("--- | --- | --- | --- | --- | --- | ---")

# Loop through each starred repository and print details
for index, repo in enumerate(all_repos, start=1):
    repo_name = repo['name']
    repo_url = repo['html_url']
    repo_description = repo['description'] if repo['description'] else "No description"
    language = repo['language'] if repo['language'] else "N/A"
    stars = repo['stargazers_count']
    forks = repo['forks_count']

    # Print each repository's details
    print(f"{index}. | {repo_name} | {repo_url} | {repo_description} | {language} | {stars} | {forks}")
```
